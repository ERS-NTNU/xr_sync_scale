```python
from ripe.atlas.cousteau import MeasurementRequest
import pandas as pd
import datetime

# 1. Define parameters: we want ping measurements
params = {
    "measurement_type": "ping",
    "start": int((datetime.datetime.utcnow() - datetime.timedelta(days=1)).timestamp()),  # 24 hours ago
    "stop": int(datetime.datetime.utcnow().timestamp()),  # now
    "target": "8.8.8.8",  # Google DNS (common, stable target)
    "limit": 1000  # limit number of results
}

# 2. Send request
is_success, results = MeasurementRequest(**params).create()

# 3. Save and structure
if is_success:
    pings = []
    for r in results['measurements']:
        if 'rtt' in r:
            pings.append({
                'timestamp': r['timestamp'],
                'probe_id': r['prb_id'],
                'rtt': r['result'][0]['rtt']  # first packet RTT
            })

    df = pd.DataFrame(pings)
    df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
    print(df.head())
else:
    print("Failed to fetch measurements!")
```

```python
from ripe.atlas.cousteau import AtlasResultsRequest
import pandas as pd
import datetime

# 1. Measurement ID (Google ping)
measurement_id = 5010

# 2. Define time window (e.g., last 1 day)
stop_time = datetime.datetime.now(datetime.timezone.utc)
start_time = stop_time - datetime.timedelta(hours=6)


# 3. Fetch results (add time filters)
kwargs = {
    "msm_id": measurement_id,
    "start": int(start_time.timestamp()),
    "stop": int(stop_time.timestamp())
}

# 4. Request
is_success, results = AtlasResultsRequest(**kwargs).create()

# 5. Parse
if is_success:
    pings = []
    for r in results:
        if 'result' in r and 'rtt' in r['result'][0]:
            pings.append({
                'timestamp': r['timestamp'],
                'probe_id': r['prb_id'],
                'rtt': r['result'][0]['rtt']
            })

    df = pd.DataFrame(pings)
    df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
    print(df.head())
else:
    print("Failed to fetch measurements!")
```

```python
from ripe.atlas.cousteau import Measurement

# 1. Instantiate the Measurement object, requesting the 'probes' field
measurement = Measurement(id=1009, optional_fields="probes")

# 2. Extract the list of probe IDs
probe_ids = [probe["id"] for probe in measurement.probes]

print("Total probes participating:", len(probe_ids))
print("Sample probe IDs:", probe_ids[:10])
```

```python
from ripe.atlas.cousteau import ProbeRequest

filters = {"tags": "NAT", "country_code": "NL", "asn_v4": "3333"}
probes = ProbeRequest(**filters)

for probe in probes:
    print(probe["id"])

# Print total count of found probes
print(probes.total_count)
```
